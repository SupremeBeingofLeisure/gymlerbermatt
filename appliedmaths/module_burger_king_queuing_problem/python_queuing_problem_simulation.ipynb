```python
import numpy as np
kubus = np.arange(1, 7, 1, dtype=int)

from queue import Queue # lets you generate a queue with maxlength

def runsim(customer=1,servants=2,minutes=20): # runs a "Warteschlangen" simulation with a queue of length 3
  q = Queue(maxsize=3)
  q.put(1) # at start t=0 there is one customer in the queue
  servings = 0
  for k in range(len(kubus)*minutes):
    throw = np.random.choice(kubus)
    if throw > len(kubus)-customer and q.qsize() < 3: # customer enters queue?
      q.put(1)
    throw = np.random.choice(kubus)
    if throw > len(kubus)-servants and q.qsize() != 0: # customer is served?
      q.get()
      servings += 1
  return servings # how many meals have been served during the simulation
```

```python
customers = 1
servants = 2
minutes = 20

l = []
for k in range(100): # runs the simulation 100 times
  l.append(runsim(customers,servants,minutes))
print(l)
```

```python
price = 10
wage = 30

for k in range(len(l)): # calculates the earnings = sales - wages
  l[k] = price * l[k] - servants * wage * minutes/60
print(l)
```

```python
sum(l)/len(l) # calculates the mean of the earnings of all simulation runs
```
