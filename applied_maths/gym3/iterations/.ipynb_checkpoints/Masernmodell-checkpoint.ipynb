```python
import numpy as np

# Anfangsbedingungen
I_0 = 20
E_0 = 30000
M_0 = 0

# Parameter
beta = 0.3 * 10**-4
iterations = 100

# Arrays zur Speicherung der Werte
I = np.zeros(iterations)
E = np.zeros(iterations)

# Anfangswerte setzen
I[0] = I_0
E[0] = E_0

# Iteration des Modells
for n in range(iterations - 1):
    I[n + 1] = beta * E[n] * I[n]
    E[n + 1] = E[n] - beta * E[n] * I[n] + 120

# 500 gleichmäßig verteilte Punkte auswählen
total_points = 500
indices = np.linspace(0, iterations - 1, total_points, dtype=int)

# Punkte für TikZ ausgeben
tikz_points = [(I[i], E[i]) for i in indices]

# TikZ-Ausgabe
print("\\begin{tikzpicture}")
print("\\draw[thick] plot coordinates {")
for x, y in tikz_points:
    print(f"({x:.4f}, {y:.4f})")
print("};")
print("\\end{tikzpicture}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Anfangsbedingungen
I_0 = 20
E_0 = 30000
M_0 = 0

# Parameter
beta = 0.3 * 10**-4
iterations = 1000

# Arrays zur Speicherung der Werte
I = np.zeros(iterations)
E = np.zeros(iterations)

# Anfangswerte setzen
I[0] = I_0
E[0] = E_0

# Iteration des Modells
for n in range(iterations - 1):
    I[n + 1] = beta * E[n] * I[n]
    E[n + 1] = E[n] - beta * E[n] * I[n] + 360

# Plot der Ergebnisse
plt.figure(figsize=(8, 5))
plt.plot(I, label='Infectious')
plt.plot(E, label='Susceptibel')
plt.xlabel('Iterations')
plt.ylabel('Population')
plt.title('Masernmodell')
plt.legend()
plt.grid(True)
plt.savefig('masernmodellpdf.pdf', format='pdf')  # Speichert den Plot als SVG-Datei
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Anfangsbedingungen
x_0 = 150
y_0 = 50
a = 0.0832
b = 0.00207
c = 0.06
d = 0.00049
time_units = 600

# Arrays zur Speicherung der Werte
x = np.zeros(time_units)
y = np.zeros(time_units)

# Anfangswerte setzen
x[0] = x_0
y[0] = y_0

# Iteration des Modells
for t in range(time_units - 1):
    x[t + 1] = x[t] * (1 + a) - b * x[t] * y[t]
    y[t + 1] = y[t] * (1 - c) + d * x[t] * y[t]

# Plot der Ergebnisse
plt.figure(figsize=(10, 5))
plt.plot(x, label='Hasen')
plt.plot(y, label='Luchse')
plt.xlabel('Zeit')
plt.ylabel('Population')
plt.title('Lotka-Volterra Modell')
plt.legend()
plt.grid(True)
plt.savefig('hasenluchsepdf.pdf', format='pdf')  # Speichert den Plot als SVG-Datei
plt.show()
```
